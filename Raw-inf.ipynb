```python
from langchain_ollama import OllamaLLM

ollama = OllamaLLM(base_url="http://localhost:11434", model="phi3")

EXAMPLE_PROMPT = """
You are an expert IPL statistician. Answer the following questions accurately and in short for each question based on IPL data:
Below is an example . 

Question: Which two teams played an IPL match on the 5th of April 2017 ?
Answer: The Indian Premier League (IPL) 2017 season commenced on April 5, 2017, with the opening match between Sunrisers Hyderabad and Royal Challengers Bangalore.
"""

questions = [
    "1. Which two teams played an IPL match on the 5th of April 2017?",
    "2. Who won the toss during the match that happened on the 7th of April 2017?",
    "3. Who won the match between KKR and RR that happened on 20/05/08?",
    "4. How many matches happened on the 9th of April 2017?",
    "5. In what locations did KKR and CSK play a match in the year 2010?",
    "6. What was the highest margin of victory (by runs) achieved in matches played at Wankhede Stadium in 2017?",
    "7. How many matches did MI play against CSK in the 2017 season, and what were the results?",
    "8. Who were the umpires for the match where the team batting second won by the smallest margin in 2017?",
    "9. Which team had the most wins in matches played at Eden Gardens across all seasons?",
    "10. What was the total number of no-results (matches without a winner) in IPL history up to the 2017 season?",
]

for i, question in enumerate(questions, 1):
    full_prompt = f"""
    {EXAMPLE_PROMPT}
    Now answer the below question
    Question: {question}
    """
    response = ollama.invoke(full_prompt)
    
    print(f"Answer to Question {i}: {response}\n")
```

```python
from langchain_ollama import OllamaLLM

ollama = OllamaLLM(base_url="http://localhost:11434", model="phi3")

EXAMPLE_PROMPT = """
You are an expert IPL statistician. Answer the following questions accurately and in short for each question based on IPL data:
Given below are few examples .
Example 1:
Question: Which two teams played an IPL match on the 5th of April 2017 ?
Answer: The Indian Premier League (IPL) 2017 season commenced on April 5, 2017, with the opening match between Sunrisers Hyderabad and Royal Challengers Bangalore.

Example 2:
Question: Who won the toss during the match that happened on the 7th of April 2017 ?
Answer: Kolkata Knight Riders won the toss and elected to field.

Example 3:
Question: Who won the match between KKR and RR that happened on 20/05/08 ?
Answer: Rajasthan Royals won by 6 wickets .

"""

questions = [
    "1. Which two teams played an IPL match on the 5th of April 2017?",
    "2. Who won the toss during the match that happened on the 7th of April 2017?",
    "3. Who won the match between KKR and RR that happened on 20/05/08?",
    "4. How many matches happened on the 9th of April 2017?",
    "5. In what locations did KKR and CSK play a match in the year 2010?",
    "6. What was the highest margin of victory (by runs) achieved in matches played at Wankhede Stadium in 2017?",
    "7. How many matches did MI play against CSK in the 2017 season, and what were the results?",
    "8. Who were the umpires for the match where the team batting second won by the smallest margin in 2017?",
    "9. Which team had the most wins in matches played at Eden Gardens across all seasons?",
    "10. What was the total number of no-results (matches without a winner) in IPL history up to the 2017 season?",
]

for i, question in enumerate(questions, 1):
    full_prompt = f"""
    {EXAMPLE_PROMPT}
    
    Now, answer the following question :

    Question: {question}
    Answer:
    """
    response = ollama.invoke(full_prompt)
    print(f"Answer to Question {i}: {response}\n")
```
